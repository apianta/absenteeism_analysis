# Now extract data for day of the week:

```python
df_reason_mod['Date'][699].weekday()
```

To apply a certain type of modification iteratively on each value from a Series or a column in a DataFrame, create a function that can execture this operation for one element, and then implement it to all values from the column of interest.

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

# Education regression:

Transform into *dummy variables*

```python
df_reason_date_mod['Education'].unique()
```

- 1 = high school
- 2 = graduate
- 3 = post graduate
- 4 = a 'master' or a 'doctor'

```python
# determine how many times each value has been counted
df_reason_date_mod['Education'].value_counts()
```

```python
# map through each variable with map() and change the 'high school' values to '0' while also changing each other value to '1'
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```
